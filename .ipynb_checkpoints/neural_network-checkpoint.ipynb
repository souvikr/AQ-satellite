```python
import pandas as pd
from sklearn import linear_model
import pandas
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
import numpy as np
import sklearn
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
df = pd.read_excel("pm_nasa.xlsx")
```

```python
df.head(5)
```

```python
df.shape
```

```python
x = pd.DataFrame(df[['TAU', 'RH','TMP','HPBL','WS10m','MONTH','LAT','LON']])
y = pd.DataFrame(df['PM24H'])
```

```python
#Keras Neural Network
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim=8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(8, input_dim=8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
seed = 1
np.random.seed(seed)
# evaluate model with standardized dataset
estimator = KerasRegressor(build_fn=baseline_model, epochs=4, batch_size=5,validation_split=0.1)
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
X_test
```

```python
estimator.fit(X_train, y_train)
pred_y = estimator.predict(X_test)
```

```python
pred_y
```

```python
kfold = KFold(n_splits=5, random_state=seed)
results = cross_val_score(estimator, x, y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
results*-1
```

```python
from sklearn.metrics import make_scorer, r2_score
fn = make_scorer(r2_score)
```

```python
kfold = KFold(n_splits=5, random_state=seed)
results = cross_val_score(estimator, x, y, cv=kfold, scoring=fn)
print("Results: %.2f (%.2f) R2 score" % (-results.mean(), results.std()))
```
