```python
import pandas as pd
from sklearn import linear_model
import pandas
from sklearn.model_selection import KFold
import numpy as np
import sklearn
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge
from scipy.stats import pearsonr
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
df = pd.read_excel("pm_nasa.xlsx")
df.head(5)
```

```python
g = set(df['StationNum'].values)
j = list(g)
df1 = df.sort_values(by=['StationNum'])
j.sort()
x = j[:75]
y = j[75:]
print(len(y))
train_station = df1[df1['StationNum'].isin(x)]
test_station = df1[df1['StationNum'].isin(y)]
train_station
```

```python
winter_train = train_station[train_station['MONTH'].isin([12,1,2])]
summer_train = train_station[train_station['MONTH'].isin([6,7,8])]

winter_test = test_station[test_station['MONTH'].isin([12,1,2])]
summer_test = test_station[test_station['MONTH'].isin([6,7,8])]

x_win = pd.DataFrame(winter_train[['TAU', 'RH','TMP','HPBL','WS10m','CLOUD']])
x_sum = pd.DataFrame(summer_train[['TAU', 'RH','TMP','HPBL','WS10m','CLOUD']])
y_sum = pd.DataFrame(summer_train['PM24H'])
y_win = pd.DataFrame(winter_train['PM24H'])

x_win_test = pd.DataFrame(winter_test[['TAU', 'RH','TMP','HPBL','WS10m','CLOUD']])
x_sum_test = pd.DataFrame(summer_test[['TAU', 'RH','TMP','HPBL','WS10m','CLOUD']])
y_sum_test = pd.DataFrame(summer_test['PM24H'])
y_win_test = pd.DataFrame(winter_test['PM24H'])
```

```python
def cross_win_sum(X,y,x_cross,y_cross,alpha):
    X_train, X_valid, y_train, y_valid = sklearn.model_selection.train_test_split(X, y, test_size = 0.2, random_state = 0)
    regression_model = Ridge(alpha)
    regression_model.fit(X_train, y_train)
    predictions_on_winter = sklearn.model_selection.cross_val_predict(regression_model,X_valid,y_valid,cv=5)
    train_error = regression_model.score(X_train, y_train)
    valid_error = regression_model.score(X_valid,y_valid)
    print("------Training-------\nTrained on: ",X_train.shape," samples |","Train Score(R2 score):",train_error)
    print("Validated on:",X_valid.shape,"samples |","Validation Score(R2 score):",valid_error)
    r2=mae=mse=rmse = 0
    r2 = pearsonr(y_valid.values.flatten(),predictions_on_winter.flatten())[0]
    mae = mean_absolute_error(y_valid,predictions_on_winter)
    mse = mean_squared_error(y_valid, predictions_on_winter)
    rmse = np.sqrt(mse)
    print("R-value: ",r2,"RMSE: ",rmse,"MAE: ",mae)
    regression_model.fit(X_train, y_train)
    predictions_on_summer = sklearn.model_selection.cross_val_predict(regression_model,x_cross,y_cross,cv=5)
    test_error = regression_model.score(x_cross, y_cross)
    print("\n---------Testing-------------\nScore(R2 score):",test_error)
    r2 = pearsonr(y_cross.values.flatten(),predictions_on_summer.flatten())[0]
    mae = mean_absolute_error(y_cross,predictions_on_summer)
    mse = mean_squared_error(y_cross,predictions_on_summer)
    rmse = np.sqrt(mse)
    print("R-value: ",r2,"RMSE: ",rmse,"MAE: ",mae)
```

```python
print("Trained on Winter - Validated on Winter - Tested on Summer")
cross_win_sum(x_win,y_win,x_sum_test,y_sum_test,0.001)

print("\n\nTrained on Winter - Validated on Winter - Tested on Winter")
cross_win_sum(x_win,y_win,x_win_test,y_win_test,0.001)

print("\n\nTrained on Summer - Validated on Summer - Tested on Summer")
cross_win_sum(x_sum,y_sum,x_sum_test,y_sum_test,0.001)

print("\n\nTrained on Summer - Validated on Summer - Tested on Winter")
cross_win_sum(x_sum,y_sum,x_win_test,y_win_test,0.0001)

```
